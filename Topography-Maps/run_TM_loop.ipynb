```python
import subprocess
import os

root_path = os.path.expanduser("~/Desktop/log_files/")
for root, dirs, files in os.walk(root_path):
    for file_name in files:
        file_path = os.path.join(root, file_name)

        # Command to execute your Python script (replace with your script path and any arguments)
        script_path = "/nfshome/natpenn/Desktop/CMS-Wafer-Testing/Topography-Maps/topography_map.py"
        command = ['python3.9', script_path, file_path]

        # Run the script as a subprocess
        process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

        # Wait for the process to complete and capture output (optional)
        stdout, stderr = process.communicate()

        # Check if there was an error
        if process.returncode != 0:
            print(f"Error running script: {stderr.decode('utf-8')}")
        else:
            print(f"Script ran successfully: {stdout.decode('utf-8')}")
```
